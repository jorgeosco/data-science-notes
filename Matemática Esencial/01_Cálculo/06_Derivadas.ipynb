# Derivadas

La derivada es una medida de cómo cambia una función a medida que su variable independiente cambia. En términos más formales, la derivada de una función en un punto dado representa la pendiente de la tangente a la curva de la función en ese punto. La derivada se denota comúnmente como $f'(x)$, $\frac{df}{dx}$.

$$ \frac{d}{dx} f(x) = f'(x) = \lim_{h \to 0} \frac{f(x + h) - f(x)}{h} $$

### Reglas de derivación

Sean $f(x)$, $g(x)$ funciones derivables, $c$ una constante y $n$ un número real.

1. $ \frac{d}{dx} [c] = 0$
2. $ \frac{d}{dx} [cx] = c\frac{d}{dx} [x]$
3. $ \frac{d}{dx} [x^n] = n \cdot x^{n-1}$
4. $ \frac{d}{dx} [e^x] = e^x$
5. $ \frac{d}{dx} [f(x) + g(x)] = f'(x) + g'(x)$
6. $ \frac{d}{dx} [f(x) - g(x)] = f'(x) - g'(x)$
7. $ \frac{d}{dx} [f(x) \cdot g(x)] = f'(x) \cdot g(x) + f(x) \cdot g'(x)$
8. $ \frac{d}{dx} \left[ \frac{f(x)}{g(x)} \right] = \frac{f'(x) \cdot g(x) - f(x) \cdot g'(x)}{[g(x)]^2}$



### Ejemplos

#### 1) Hallar la derivada de $f(x) = 3x^2 + 2x - 5$ y evaluar en $x = 2$.

```python
x = symbols('x')
f = 3*x**2 + 2*x - 5
derivada = diff(f, x)
derivada
```

Como se observa, la derivada de $f(x)$ es $f'(x) = 6x + 2$, con esto podemos escribir la función en código

Así, evaluar la derivada en $x = 2$

```python
def f(x):
    return 3*x**2 + 2*x - 5
def dx_f(x):
    return 6*x + 2
```

```python
x = 2
dx_f(x)
```

```python
x = symbols('x')
f = sin(x)
derivada = diff(f, x)
derivada
```

Como se observa, la derivada de $f(x)$ es $f'(x) = \cos(x)$, con esto podemos escribir la función en código

## Derivadas Parciales

La derivada parcial es una extensión del concepto de derivada a funciones de varias variables. En lugar de medir cómo cambia una función con respecto a una sola variable, la derivada parcial mide cómo cambia la función con respecto a una de sus variables mientras se mantienen las otras variables constantes.

La notación común para la derivada parcial de una función $f(x, y)$ con respecto a $x$ es $\frac{\partial f}{\partial x}$ o $f_x$. De manera similar, la derivada parcial con respecto a $y$ se denota como $\frac{\partial f}{\partial y}$ o $f_y$.

En una variable la derivada es la pendiente de la tangente a la curva de la función en un punto dado. En varias variables, la derivada parcial mide la pendiente de la función en una dirección específica, manteniendo las otras variables constantes. Tecnicamente, estas pendientes se las denominan gradientes.

### Ejemplos

#### 1) Hallar la derivada parcial de $f(x, y) = x^3 + y^2$ con respecto a $x$ y $y$ y evaluar en $(1,2)$.

```python
x, y = symbols('x y')
f = x**3 + y**2
```

```python
plot3d(f)
```

Hallamos la derivada parcial con respecto a $x$.

```python
dx_f = diff(f, x)
dx_f
```

Hallamos la derivada parcial con respecto a $y$.

```python
dy_f = diff(f, y)
dy_f
```

Como se observa, la derivada parcial de $f(x,y)$ con respecto a $x$ es $f_x(x,y) = 3x^2$ y con respecto a $y$ es $f_y(x,y) = 2y$. Con esto podemos escribir las funciones en código

Así, evaluar la derivada parcial en $(1,2)$

```python
def f(x, y):
    return x**3 + y**2
def dx_f(x, y):
    return 3*x**2
def dy_f(x, y):
    return 2*y
```

```python
x = 1
y = 2
dx_f(x, y), dy_f(x, y)
```

## Regla de la cadena

La regla de la cadena es una fórmula utilizada para calcular la derivada de una función compuesta. Si tenemos dos funciones $f(x)$ y $g(x)$, la regla de la cadena nos permite encontrar la derivada de la función compuesta $f(g(x))$.
La regla de la cadena se expresa como:
$$ \frac{d}{dx} [f(g(x))] = f'(g(x)) \cdot g'(x) $$

### Ejemplos

#### 1) Hallar la derivada de $f(x) = \sin(3x^2 + 2)$.

```python
x = symbols('x')
f = sin(3*x**2 + 2)
derivada = diff(f, x)
derivada
```

Resolviendo con la regla de la cadena paso a paso.

Definimos $f(x)$ en funciones separadas.
   - $y = 3x^2 + 2$
   - $z = \sin(y)$

```python
x, y = symbols('x y')
_y = 3*x**2 + 2
z = sin(y)
```

Hallamos la derivada de cada una.
   - $ \frac{dy}{dx} = 6x$
   - $ \frac{dz}{dy} = \cos(y)$

```python
dy_dx = diff(_y)
dy_dx
```

```python
dz_dy = diff(z)
dz_dy
```

Aplicamos el producto de las derivadas.
- $ \frac{dz}{dy} \cdot \frac{dy}{dx} = \cos(y) \cdot 6x$

Sustituimos $y$ en la ecuación.
- $ \cos(y) \cdot 6x = \cos(3x^2 + 2) \cdot 6x$

Finalmente, la derivada de $f(x)$ es:
- $ \frac{d}{dx} [f(x)] = 6x \cdot \cos(3x^2 + 2) $

```python
df_dx = (dy_dx * dz_dy).subs(y, _y)
df_dx
```

#### 2) Hallar la derivada de $f(x) = e^{3x^2 + 2}$.

```python
x = symbols('x')
f = exp(3*x**2 + 2)
derivada = diff(f, x)
derivada
```

```python
dy_dx = diff(_y)
dy_dx
```

```python
dz_dy = diff(z)
dz_dy
```

Aplicamos el producto de las derivadas.
- $ \frac{dz}{dy} \cdot \frac{dy}{dx} = e^{y} \cdot 6x$
Sustituimos $y$ en la ecuación.
- $ e^{y} \cdot 6x = e^{3x^2 + 2} \cdot 6x$

```python
df_dx = (dy_dx * dz_dy).subs(y, _y)
df_dx
```
