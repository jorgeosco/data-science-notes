```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# Funciones

Las funciones son expresiones que definen relaciones entre dos o más variables.

## Función de una variable

Una función es una regla de correspondensia que se asigna a cada elemento de un conjunt A un único elemento de otro conjunto B. Se denota como $f: A \to B$. Esto significa que dado un elemento $x$ de A, existe un único elemento $y$ de B tal que $f(x) = y$.

- Dominio: Conjunto de valores que toma la variable independiente $x$.
- Codominio: Conjunto de valores que toma la variable dependiente $y$.
- Rango: Valores del Codominio que son imagenes de $x$, es decir $f(x)$.

#### Ejemplo: $ f(x) = 2x + 1 $

```python
def f(x):
    return 2 * x + 1
```

```python
x_values = [0, 1, 2, 3]
y_values = [f(x) for x in x_values]
```

```python
df = pd.DataFrame({'x': x_values, 'f(x)': y_values})
df
```

```python
plt.plot(df['x'], df['f(x)'], marker='o', label='f(x) = 2x + 1')
plt.title(r'$f(x) = 2x + 1$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
x = np.linspace(-10, 10, 20)
y = f(x)
```

```python
plt.plot(x, y, label='f(x) = 2x + 1', color='blue')
plt.title('f(x) = 2x + 1')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.axhline(0, color='black', linewidth=2, linestyle='-')
plt.axvline(0, color='black', linewidth=2, linestyle='-')
plt.grid(True)
plt.legend()
plt.show()
```

#### Ejemplo: $ f(x) = x^2 + 3 $

```python
def f(x):
    return x**2 + 3
```

```python
x = np.linspace(-10, 10, 100)
y = f(x)
```

```python
plt.plot(x, y, label='f(x) = x^2 + 3', color='green')
plt.title(r'$f(x) = x^2 + 3$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.axhline(0, color='black', linewidth=2, linestyle='-')
plt.axvline(0, color='black', linewidth=2, linestyle='-')
plt.grid(True)
plt.legend()
plt.show()
```

## Funciones de dos variables

Las funciones de dos variables son expresiones que dependen de dos variables independientes. Se denota como $f: A \times B \to C$.

Esto significa que dado un par de elementos $(x, y)$ de A y B, existe un único elemento $z$ de C tal que $f(x, y) = z$.

#### Ejemplo: $ f(x, y) = x + 2y $

```python
def f(x, y):
    return x + (2 * y)
```

```python
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
fig = plt.figure(figsize=(11, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')
ax.set_title(r'$f(x, y) = x + 2y$')
ax.set_xlabel('eje x')
ax.set_ylabel('eje y')
ax.set_zlabel(r'$f(x, y)$')
plt.show()
```

#### Ejemplo: $ f(x, y) = x^2 + y^2 $

```python
def f(x, y):
    return x**2 + y**2
```

```python
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
fig = plt.figure(figsize=(11, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')
ax.set_title(r'$f(x, y) = x^2 + y^2$')
ax.set_xlabel('eje x')
ax.set_ylabel('eje y')
ax.set_zlabel(r'$f(x, y)$')
plt.show()
```

## Nota

Las funciones de n variables son expresiones que dependen de n variables independientes. Se denota como $f: A_1 \times A_2 \times ... \times A_n \to C$.
Esto significa que dado un vector de elementos $(x_1, x_2, ..., x_n)$ de $A_1, A_2, ..., A_n$, existe un único elemento $z$ de C tal que $f(x_1, x_2, ..., x_n) = z$.

Ya sea que tengamos $1$ o $n$ variables independientes, solo hay una variable dependiente.
